```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# Read the dataset into a Pandas DataFrame
prosper_loan_df = pd.read_csv("prosper_loan_data.csv")
```

## Distribution Borrower's Annual Percentage Rate  

> The Borrower's Annual Percentage Rate started with a low APR of 0.05% and gradually increase to 0.10%, afterwards, it maintained almost a steady APR between 0.15% and 0.30% before it attained its peak count at 0.35%.

```python
#  BorrowerAPR distribution 
plt.figure(figsize=[6,4])
bins_base = np.arange(0, prosper_loan_df['BorrowerAPR'].max(), 0.005)
plt.hist(data=prosper_loan_df, x = 'BorrowerAPR', bins = bins_base)
plt.title("Distribution of Borrower's Annual Percentage Rate")
plt.xlabel("Borrower's Annual Percentage Rate (%)")
plt.ylabel("Counts")
plt.xticks(np.arange(0, prosper_loan_df['BorrowerAPR'].max()+0.05, 0.05));
```

## Relationship Between Borrower's Annual Percentage Rate and Lender Yield

> A strong postive correlation can be oberved between the Borrower's Annual Percentage Rate and Lender Yield, because as the Borrower's Annual Percentage Rate increases, the Lender Yield also increase.

```python
# Relationship Between BorrowerAPR and LenderYield
plt.figure(figsize=[6,4])
sb.regplot(data=prosper_loan_df,x="LenderYield",y="BorrowerAPR",scatter_kws={"alpha":0.02});
plt.xlabel("Borrower's Annual Percentage Rate (%)");
plt.ylabel("Lender Yield (%)");
plt.title("Relationship Between Borrower APR  and Lender Yield");
```

## Relationship Between Borrower's Annual Percentage Rate and Prosper Score

> There is a custom risk score built using Prosper data historical . The score ranges from 1-10, with 10 being the best, or lowest risk score. The plot below depicts that the higher the Borrower's Annual Percentage Rate, the lower the Prosper Score. This also translate to a negative correlation between BorrowerAPR and ProsperScore. This implies that borrowers with higher rating have the chances to receive lower interest rate and APR.

```python
plt.figure(figsize = [10, 5])
sb.regplot(data=prosper_loan_df,x="BorrowerAPR", y="ProsperScore");
plt.xlabel("Borrower's Annual Percentage Rate (%)");
plt.ylabel("Prosper Score (%)");
plt.title("Relationship Between Borrower's Annual Percentage Rate and Prosper Score");
```
